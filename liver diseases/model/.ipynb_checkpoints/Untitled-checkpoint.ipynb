```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train[ : 5]
```

```python
import tensorflow as tf
```

```python
ANN_model = tf.keras.models.Sequential()
```

```python
ANN_model.add(tf.keras.layers.Dense(64, input_dim = 11, activation = 'relu'))
ANN_model.add(tf.keras.layers.Dense(128, activation = 'relu'))
ANN_model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
ANN_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ANN_model.fit(x_train, y_train, batch_size = 1, epochs = 200)
```

```python
loss, acc_ann = ANN_model.evaluate(x_train, y_train, verbose = 0)
```

```python
print(acc_ann)
```

```python
y_pred = ANN_model.predict(x_test)
y_pred[y_pred <= 0.5] = 0
y_pred[y_pred > 0.5] = 1
```
